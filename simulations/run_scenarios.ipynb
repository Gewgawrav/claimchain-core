```python
parsed_logs_folder = 'Enron/parsing/'
social_graph = pickle.load(open(parsed_logs_folder + "social.pkl", "rb"))
log = pickle.load(open(parsed_logs_folder + "replay_log.pkl", "rb"))
```

```python
def visualize_propagation(scenario_name, kp, hp):
    kp.plot.area();
    plt.title("%s - Encryption keys propagation" % scenario_name)
    plt.xlabel("Emails sent")
    plt.ylabel("Link Key status")

    hp.plot.area();
    plt.title("%s - Head propagation" % scenario_name)
    plt.xlabel("Emails sent")
    plt.ylabel("Link Head status")

    plt.show()
```

```python
kp, hp = simulate_static_autocrypt(social_graph, log)
```

```python
visualize_propagation('Static Autocrypt', kp, hp)
```

```python
kp, hp = simulate_static_claimchain_no_privacy(social_graph, log)
```

```python
visualize_propagation('Claimchain without privacy', kp, hp)
```

```python
kp, hp = simulate_static_claimchain_with_privacy(social_graph, log)
```

```python
visualize_propagation('Claimchain with privacy', kp, hp)
```
